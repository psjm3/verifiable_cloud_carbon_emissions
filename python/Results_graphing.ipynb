```python
import matplotlib
import matplotlib.pyplot as plt

num_of_leaf = [2**8, 2**10, 2**12, 2**14, 2**16, 2**18]

time_tree_gen = [1, 6, 27, 120, 535, 2371]
tree_time_fig, tree_time_ax = plt.subplots()
tree_time_ax.plot(num_of_leaf, time_tree_gen, linestyle='-', marker='o')
tree_time_ax.set_xscale('log', base=2)
tree_time_ax.set_yscale('log', base=2)
tree_time_ax.set_xticks(num_of_leaf)
tree_time_ax.get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
tree_time_ax.set_xlabel("Number of customer leaf nodes")        # Label for the X-axis
tree_time_ax.set_ylabel("Time ($log_2$ seconds)")        # Label for the Y-axis
tree_time_ax.set_title("Time taken to build Merkle Tree in memory")  # Chart title
plt.savefig('time_tree_gen.png', dpi=500)
plt.show()

time_customer_shares = [218, 585, 1317, 4174, 15481, 58821]
time_customer_shares_fig, time_customer_shares_ax = plt.subplots()
time_customer_shares_ax.plot(num_of_leaf, time_customer_shares, linestyle='-', marker='o')
time_customer_shares_ax.set_xscale('log', base=2)
time_customer_shares_ax.set_yscale('log', base=2)
time_customer_shares_ax.set_xticks(num_of_leaf)
time_customer_shares_ax.get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
time_customer_shares_ax.set_xlabel("Number of customer leaf nodes")        # Label for the X-axis
time_customer_shares_ax.set_ylabel("Time ($log_2$ seconds)")        # Label for the Y-axis
time_customer_shares_ax.set_title("Time taken to complete the customer shares proofs for a set of tree sizes")  # Chart title
plt.savefig('time_customer_shares.png', dpi=500)
plt.show()

time_overall = [1603, 1980, 2696, 5552, 16920, 39600]
time_overall_fig, time_overall_ax = plt.subplots()
time_overall_ax.plot(num_of_leaf, time_overall, linestyle='-', marker='o')
time_overall_ax.set_xscale('log', base=2)
time_overall_ax.set_yscale('log', base=2)
time_overall_ax.set_xticks(num_of_leaf)
time_overall_ax.get_yaxis().set_major_formatter(matplotlib.ticker.ScalarFormatter())
time_overall_ax.set_xlabel("Number of customer leaf nodes")        # Label for the X-axis
time_overall_ax.set_ylabel("Time ($log_2$ seconds)")        # Label for the Y-axis
time_overall_ax.set_title("Time taken to complete all the proofs for one customers")  # Chart title
plt.savefig('time_overall.png', dpi=500)
plt.show()
```
